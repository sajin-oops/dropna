```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Line Plot
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.plot(x, y, marker='o')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Bar Chart
plt.subplot(2, 2, 2)
plt.bar(x, y, color='orange')
plt.title('Bar Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Scatter Plot
plt.subplot(2, 2, 3)
plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Pie Chart
labels = ['A', 'B', 'C', 'D', 'E']
plt.subplot(2, 2, 4)
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
